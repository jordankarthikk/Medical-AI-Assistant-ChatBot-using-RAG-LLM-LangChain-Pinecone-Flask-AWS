```python
from langchain_community.document_loaders import PyPDFLoader,DirectoryLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
def load_pdf_files(data):
    load=DirectoryLoader(data
                        ,glob="*.pdf"
                        ,loader_cls=PyPDFLoader)
    
    documents= load.load()
    return documents
```

```python
extracted_doc=load_pdf_files("../data")
```

```python
len(extracted_doc)
```

```python
def filter_to_minimal_doc(docs:list[Document]) ->list[Document]:
    
    minimal_doc: list[Document] = []
    for doc in docs:
        src=doc.metadata.get("source")
        minimal_doc.append(
                         Document(page_content=doc.page_content
                         ,metadata={"source":src}))
    return minimal_doc
```

```python
minimal_docs = filter_to_minimal_doc(extracted_doc)
```

```python
#split the document into smaller chunks
def text_split(minimal_doc):
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=500,
        chunk_overlap=20,
    )
    return text_splitter.split_documents(minimal_doc)


text_chunks = text_split(minimal_docs)
```
